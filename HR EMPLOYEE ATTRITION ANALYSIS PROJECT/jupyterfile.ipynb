```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
print("First 5 Rows")
display(df.head())

print("\nShape of Dataset")
print(df.shape)

print("\nColumn Names")
print(df.columns)

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
display(df.describe())
```

```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Count")

plt.show()
```

```python
attrition_percentage = df['Attrition'].value_counts(normalize=True) * 100

print("\nAttrition Percentage")
print(attrition_percentage)
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=20, kde=True)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(x='Attrition', y='Age', data=df)

plt.title("Age vs Attrition")

plt.show()
```

```python

plt.figure(figsize=(8,5))

sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)

plt.title("Monthly Income vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='OverTime', hue='Attrition', data=df)

plt.title("Overtime vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Department', hue='Attrition', data=df)

plt.title("Department Wise Attrition")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(x='JobRole', hue='Attrition', data=df)

plt.title("Job Role Wise Attrition")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)

plt.title("Job Satisfaction vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x='MaritalStatus', hue='Attrition', data=df)

plt.title("Marital Status vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x='EducationField', hue='Attrition', data=df)

plt.title("Education Field vs Attrition")

plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)

plt.title("Years At Company vs Attrition")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)

plt.title("Work Life Balance vs Attrition")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(16,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    fmt='.1f',
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("\n================ TOP INSIGHTS ================\n")

# Attrition Count
print("Attrition Count:")
print(df['Attrition'].value_counts())

# Department Wise Attrition
print("\nDepartment Wise Attrition:")
print(pd.crosstab(df['Department'], df['Attrition']))

# Overtime vs Attrition
print("\nOvertime vs Attrition:")
print(pd.crosstab(df['OverTime'], df['Attrition']))

# Job Role vs Attrition
print("\nJob Role vs Attrition:")
print(pd.crosstab(df['JobRole'], df['Attrition']))
```

```python
df.to_csv("cleaned_hr_attrition.csv", index=False)

print("\nCleaned dataset saved successfully!")
```
